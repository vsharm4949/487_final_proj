## Task 1

```python
from simpletransformers.classification import ClassificationModel

# define hyperparameter
train_args1 ={"reprocess_input_data": True,
             "fp16":False,
             "num_train_epochs": 10,
             "overwrite_output_dir": True
            }

# Create a ClassificationModel
model1 = ClassificationModel(
    "bert", "bert-base-chinese",
    num_labels=2,
    args=train_args1
)
```

```python
df_train1 = pd.read_csv("/content/drive/Shareddrives/NLP Final Project/Dataset/balanced_train_task1.csv")
df_test1 = pd.read_csv("/content/drive/Shareddrives/NLP Final Project/Dataset/balanced_test_task1.csv")

df_train1.rename(columns={"joke": "text"}, inplace=True)
df_test1.rename(columns={"joke": "text"}, inplace=True)

df_train1.drop(columns=['id', "Unnamed: 0"], inplace=True)
df_test1.drop(columns=['id', "Unnamed: 0"], inplace=True)

df_train1.head()
```

```python
model1.train_model(df_train1) # TODO: Retrain with 10 epochs
```

```python
from sklearn.metrics import f1_score, accuracy_score


def f1_multiclass(labels, preds):
    return f1_score(labels, preds, average='macro')
    
result1, model_outputs1, wrong_predictions1 = model1.eval_model(df_test1, f1=f1_multiclass, acc=accuracy_score)

print(result1)
```

```python
torch.save(model1, 'roberta-task1.clf') # Remeber to move file to drive
```
